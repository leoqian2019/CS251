```python
degree = 7
fit.poly_regression("X","Y",degree)
check.poly_regression("X","Y",degree)
check.initialize("X","Y",fit.get_fitted_slope(),fit.get_fitted_intercept(),degree)

y_pred = check.get_ypred()
check.R2 = check.r_squared(y_pred)

fit.scatter("X","Y","X/Y training")
plt.figure()
check.scatter("X","Y","X/Y validation")


print("MSE training:",fit.m_sse)
residuals = check.compute_residuals(y_pred)
m_sse = np.mean( residuals**2 )
print("MSE validation:",m_sse)
```

```python
poly_data = data.Data("data/Literacy_PPP.csv")
lin_reg = linear_regression.LinearRegression(poly_data)

lin_reg.poly_regression("PPP", "Literacy_rate",3)
lin_reg.scatter("PPP", "Literacy_rate", "PPP vs. Literacy_rate")
plt.xlabel("GDP per Capita")
plt.ylabel("literacy level")
```

After running the polynomial regression on the dataset comparing gdp per capita (PPP) and literacy level, I found out that there's a positive relationship between PPP and literacy level and that the degree of power is between 2 and 3

P.S data describes 35 countries in 2019 with different ppp value and literacy level

But in order to test if the data overfits, I have to separate the data into two groups, which first half is the traning dataset and the second half is the validation dataset.

```python
poly_data1 = data.Data("data/Literacy_PPP.csv")
poly_data2 = data.Data("data/Literacy_PPP.csv")


fit = linear_regression.LinearRegression(poly_data1)
check = linear_regression.LinearRegression(poly_data2)
np.random.seed(0)
fit.data.permutation()
check.data.data = fit.data.data[18:36].copy()
fit.data.limit_samples(0, 18)


degree = 1
fit.poly_regression("PPP","Literacy_rate",degree)
check.poly_regression("PPP","Literacy_rate",degree)
check.initialize("PPP","Literacy_rate",fit.get_fitted_slope(),fit.get_fitted_intercept(),degree)

y_pred = check.get_ypred()
check.R2 = check.r_squared(y_pred)

fit.scatter("PPP","Literacy_rate","PPP/Literacy_rate training")
plt.figure()
check.scatter("PPP","Literacy_rate","PPP/Literacy_rate validation")


print("MSE training:",fit.m_sse)
residuals = check.compute_residuals(y_pred)
m_sse = np.mean( residuals**2 )
print("MSE validation:",m_sse)
```

After testing, degree over 1 would give a significant overfit since the r2 for degree of 2 or 3 for the validation dataset is much lower than the one for the traning dataset. In this case, it would be normal to assume that the degree of best fit here is 1

It's also interesting to see if swaping the independent variable and the dependent variable would still show linear relationship

```python
poly_data = data.Data("data/Literacy_PPP.csv")
lin_reg = linear_regression.LinearRegression(poly_data)

lin_reg.poly_regression("Literacy_rate", "PPP",1)
lin_reg.scatter("Literacy_rate", "PPP", "Literacy_rate vs. PPP")
plt.xlabel("literacy level")
plt.ylabel("ppp")
```

Interstingly enough that we can still see some sort of linear relationship here, so it's either that one of these two features are independent variable or that there are other variables that affect both of them at the same time.
